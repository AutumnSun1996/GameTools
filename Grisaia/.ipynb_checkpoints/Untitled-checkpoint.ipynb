```python
import pyquery
import re
```

```python
pq = pyquery.PyQuery(open(r"D:\Games\GameRoutes\Grisaia\灰色的果实.html", 'r', -1, "UTF-8").read())
pq
```

```python
lines = []
pos = 0
top = re.compile(r"top: ([\.\d]+)")
for item in pq("div.textLayer > div").items():
    top_ = top.search(item.attr("style"))
    if not top_:
        print(top, item)
        continue
    cur = float(top_.group(1))
    print(item, cur)
    if abs(cur - pos) > 10:
        print(pos)
        lines.append("")
        pos = cur
    lines[-1] += item.text()
lines
```

```python
print("\n".join(lines))
```
