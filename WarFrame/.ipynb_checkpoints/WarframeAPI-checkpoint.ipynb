```python
def check(content):
    for item in re.finditer(b"([./\w\d=]+)", content):
        res = item.group(1)
        start = item.span()[0]
        if len(res) < 5:
            print("    ", start, res, content[start-5:start+10])
        else:
            print(start, res)
```
